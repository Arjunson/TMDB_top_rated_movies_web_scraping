```python
df = pd.DataFrame()
```

```python
# At the time of making this project there were 503 pages for top rated movies


for i in range(1,50):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=943dc6acaaa1d9b5c10444b8f11dcfac&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)
    
```
